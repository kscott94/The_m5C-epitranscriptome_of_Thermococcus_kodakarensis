```r
color_by_reg <- function(fc){
    if(fc >= 1){return('gain')}
    else if(fc <= -1){return("loss")}
    else return('nochange')
}

n_by_reg <- function(d){
    up <- nrow(d %>% filter(log2FC>=1) )
    down <- nrow(d %>% filter(log2FC<= -1) )
    unchagned <- nrow(d %>% filter(log2FC <1 & log2FC> -1) )
        
    return(c(up,down,unchagned))
}

clean_text <- function(value, input, output){
    if(value == input){return(output)}
    else{return(value)}
    
}

clean_text("antisense-mRNA", "antisense-mRNA", "a-mRNA")
```

```r
TK0360 <- data.frame(rbind(TK0360exoS, TK0360statS)) %>% mutate(strain='TK0360') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC)) %>% 
        filter(difference == "ABS LOSS" | difference == "ABS GAIN")

TK0872 <- data.frame(rbind(TK0872exoS, TK0872statS)) %>% mutate(strain='TK0872') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC)) 

TK1935 <- data.frame(rbind(TK1935exoS, TK1935statS)) %>% mutate(strain='TK1935') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC)) 

TK2122 <- data.frame(rbind(TK2122exoS, TK2122statS)) %>% mutate(strain='TK2122') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC)) 

TK2304 <- data.frame(rbind(TK2304exoS, TK2304statS)) %>% mutate(strain='TK2304') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC)) 

TK2304_1935 <- data.frame(rbind(TK2304_1935exoS, TK2304_1935statS)) %>% mutate(strain='TK2304_1935') %>% 
        rowwise %>% mutate(reg=color_by_reg(log2FC))
```
