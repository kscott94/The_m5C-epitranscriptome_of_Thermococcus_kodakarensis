```r
#####HELPERS 

load_cgmap <- function(cgmap,strain, replicate, growth_phase, metabolic_condition='S'){
    cols = c('chromosome', 'base', 'position', 'trinuc_context','dinuc_context', 
         "methylation_frequency", "count_mC", "coverage")
    outdata <- read.delim(cgmap, sep='\t', header=FALSE,col.names=cols) %>%
        mutate(strain=strain, growth_phase=growth_phase, 
               metabolic_condition=metabolic_condition, replicate=replicate)

}


#clean up raw CGmaps
processCGmap <- function(cgmap){
    strain = unique(cgmap$strain)
    rep = unique(cgmap$replicate)
    
    final_data <- cgmap %>% 
        dplyr::select(-trinuc_context, -dinuc_context) %>%
        dplyr::rename(!!paste("coverage","_",strain,"_",rep, sep="") := coverage) %>%
        dplyr::rename(!!paste("mC_count","_",strain,"_",rep, sep="") := count_mC) %>%
        dplyr::rename(!!paste("methylation_frequency","_",strain,"_",rep, sep="") := methylation_frequency) %>%
        dplyr::select(-replicate, -strain)
    return(final_data)
}

# get standard deviation for 3 reps
get_sd <- function(f1,f2,f3,c1,c2,c3,min_cov){
    ls <-c()
    if(c1>=min_cov){
        ls <- c(ls,f1)
    }
    if(c2>=min_cov){
        ls <- c(ls, f2)

    }
    
    if(c3>=min_cov){
       ls <- c(ls, f3)

    }
    if (length(ls>1)){
        ans = round(sd(ls),3)
    } else{
        ans = NA
    }
    return(ans)
}


# get standard deviation for 2 reps
get_sd_2reps <- function(f1,f2,c1,c2,min_cov=47){
    ls <-c()
    if(c1>=min_cov){
        ls <- c(ls,f1)
    }
    if(c2>=min_cov){
        ls <- c(ls, f2)
    }
    if (length(ls>1)){
        ans = round(sd(ls),3)
    } else{
        ans = NA
    }
    return(ans)
}


#merge data for replicate deletion strains
merge_del_reps <- function(x,y, strain){
    cov_1 <- paste("coverage_",strain, "_Dec2016", sep='')
    cov_2 <- paste("coverage_",strain, "_Oct2020", sep='')

    d <- x %>%
        merge(y=y, 
              by=c("chromosome","position", 'reference_nucleotide', 'growth_phase','metabolic_condition'), 
              all=TRUE) %>%
    
            mutate(!!as.name(cov_1) := replace_na(!!as.name(cov_1),0)) %>%
            mutate(!!as.name(cov_2) := replace_na(!!as.name(cov_2),0))
    
    return(d)
}


# determine if high confidence, reproducible sites with from 3 reps
detect_hiconf_3reps <- function(f1,m1,p1,c1,f2,m2,p2,c2,f3,m3,p3,c3, min_cov=47,
                                min_freq=.10, min_freq2 = 0.05) {
    ls <- c()
    
    if(c1 == 'NA' ) {c1<-0}
    if(c2 == 'NA' ) {c2<-0}
    if(c3 == 'NA' ) {c3<-0}
    if(c1 < min_cov & c2 < min_cov & c3 < min_cov) {return(NA)}
    
    else{
    
    
    if(c1<min_cov){ls <- c(ls,"NA")}
    else if(  (c1>=min_cov & m1>=p1 & f1>=min_freq) | (m1>50 & f1>=min_freq2) ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    if(c2<min_cov){ls <- c(ls,"NA")}
    else if( (c2>=min_cov & m2>=p2 & f2>=min_freq) | (m2>50 & f2>=min_freq2) ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    if(c3<min_cov){ls <- c(ls,"NA")}
    else if( (c3>=min_cov & m3>=p3 & f3>=min_freq)| (m3>50 & f3>=min_freq2) )  {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    ans_t <- length(which(ls==TRUE))
    ans_na <- length(which(ls=="NA"))
    ans_f <- length(which(ls==FALSE))

    if(ans_na >= 3){return(NA)}
    if(ans_na == 2 & ans_t == 1){return(NA)}
    else if(ans_t >= 2){return(TRUE)}
    else { return(FALSE)}
        }
}



# determine if high confidence, reproducible sites with from 2 reps
detect_hiconf_2reps <- function(f1,m1,p1,c1,f2,m2,p2,c2, min_cov=47,
                                min_freq=.10, min_freq2 = 0.05) {
    ls <- c()
    
    if(c1 == 'NA' ) {c1<-0}
    if(c2 == 'NA' ) {c2<-0}
    
    if(c1 < min_cov & c2 < min_cov) {return(NA)}
    else{
    
    
    if(c1<min_cov){ls <- c(ls,"NA")}
    else if(  (c1>=min_cov & m1>=p1 & f1>=min_freq) | (m1>50 & f1>=min_freq2) ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    if(c2<min_cov){ls <- c(ls,"NA")}
    else if( (c2>=min_cov & m2>=p2 & f2>=min_freq) | (m2>50 & f2>=min_freq2) ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}

    ans_t <- length(which(ls==TRUE))
    ans_na <- length(which(ls=="NA"))
    ans_f <- length(which(ls==FALSE))

    if(ans_f >= 1  ){return(FALSE)}
    else if(ans_t >= 2){return(TRUE)}
    else if(ans_t == 1 & ans_na ==1){return(TRUE)}
    else { return(FALSE)}
        
        }
}

    
    
# determine if high confidence, reproducible sites with from 1 rep
detect_hiconf_1rep <- function(f1,m1,p1,c1, min_cov=47, min_freq=.10,min_freq2=0.05) {
   
    if(c1 == 'NA') {c1<-0}
    
    if(c1 < min_cov) {return(NA)}
    else if( (c1>=min_cov & m1>=p1 & f1>=min_freq) | (m1>50 & f1>=min_freq2) ) {return(TRUE)}
    else {return(FALSE)}
    
}

# determine if high confidence in all three 3 reps
super_hiconf_3reps <- function(f1,m1,p1,c1,f2,m2,p2,c2,f3,m3,p3,c3, min_cov=47,
                                min_freq=.10, min_freq2 = 0.05) {
    ls <- c()
    
    if(c1 == 'NA' ) {c1<-0}
    if(c2 == 'NA' ) {c2<-0}
    if(c3 == 'NA' ) {c3<-0}

    
    if(  (c1>=min_cov & m1>=p1 & f1>=min_freq) | (c1>1000 & f1>=min_freq2)  ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    
    if( (c2>=min_cov & m2>=p2 & f2>=min_freq) | (c2>=1000 & f2>=min_freq2) ) {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    if( (c3>=min_cov & m3>=p3 & f3>=min_freq) | (c3>=1000 & f3>=min_freq2) )  {ls <- c(ls,TRUE)}
    else {ls <- c(ls,FALSE)}
    
    ans <- length(which(ls==TRUE))
    
    if(ans >= 3){return(TRUE)}
    else { return(FALSE)}
}



#compare standard deviations
rm_overlap <- function(sp, ssd, cp,csd){
    ssd <- ssd*2
    csd <- csd*2

    r1 <- c(sp-ssd, sp+ssd)
    r2 <- c(cp-csd, cp+csd)
    ans <- r1 %overlaps% r2
    return(ans)
    
}


# get p value of each m5C site using binomial distribution
get_pvalue <- function(s1_mC,s1_cov, s2_mC,s2_cov,
                       c1_mC,c1_cov,c2_mC,c2_cov,c3_mC,c3_cov){
    
    if(s1_cov == 0 & s2_cov == 0){return(NA)} 
    
    else{

    s1_mC = s1_mC +1 
    s1_cov =  s1_cov +1 
    s2_mC = s2_mC +1 
    s2_cov = s2_cov +1 
    c1_mC = c1_mC +1 
    c1_cov = c1_cov +1 
    c2_mC = c2_mC +1 
    c2_cov = c2_cov +1 
    c3_mC = c3_mC +1 
    c3_cov = c3_cov +1 
    
    
    treatment <- c('sample','sample','control','control','control')
    success <- c(s1_mC,s2_mC,c1_mC,c2_mC,c3_mC)
    failure <- c(s1_cov-s1_mC,s2_cov-s2_mC,c1_cov-c1_mC,c2_cov-c2_mC,c3_cov-c3_mC)
    da <- data.frame(treatment, success,failure)

    fit <- glm(cbind(da$success,da$failure) ~ treatment, family = binomial, data = da)
    p_val <- summary(fit)$coefficients[8]
    
    return(p_val) }
}

get_pvalue_1rep <- function(s1_mC,s1_cov,
                       c1_mC,c1_cov,c2_mC,c2_cov,c3_mC,c3_cov){
    
    if(s1_cov == 0){return(NA)}
    
    else{

    s1_mC = s1_mC +1 
    s1_cov =  s1_cov +1 
    c1_mC = c1_mC +1 
    c1_cov = c1_cov +1 
    c2_mC = c2_mC +1 
    c2_cov = c2_cov +1 
    c3_mC = c3_mC +1 
    c3_cov = c3_cov +1 
    
    treatment <- c('sample','control','control','control')
    success <- c(s1_mC,c1_mC,c2_mC,c3_mC)
    failure <- c(s1_cov-s1_mC,c1_cov-c1_mC,c2_cov-c2_mC,c3_cov-c3_mC)
    da <- data.frame(treatment, success,failure) 
    
    fit <- glm(cbind(da$success,da$failure) ~ treatment, family = binomial, data = da)
    p_val <- summary(fit)$coefficients[8]
    return(p_val) }
}

CompareCGmap <- function(strain_cgmap, TS559_cgmap, strain, minimum_freq=.10, min_cov=47){
    
    meth_freq_r1 <- paste("methylation_frequency_", strain, "_Dec2016", sep="")
    m5C_count_r1 <- paste("mC_count_", strain, "_Dec2016", sep="")
    cov_r1 <- paste("coverage_", strain, "_Dec2016", sep="")

    meth_freq_r2 <- paste("methylation_frequency_", strain, "_Oct2020", sep="")
    m5C_count_r2 <- paste("mC_count_", strain, "_Oct2020", sep="")
    cov_r2 <- paste("coverage_", strain, "_Oct2020", sep="")
    
    
    #get quantile values for m5C count
    q_TS559_march <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_March2016>=min_cov, mC_count_TS559_March2016>=0))$mC_count_TS559_March2016, .99)[[1]]
    
    q_TS559_dec <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_Dec2016>=min_cov, mC_count_TS559_Dec2016>=0))$mC_count_TS559_Dec2016, .99)[[1]]
    
    q_TS559_oct <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_Oct2020>=min_cov, mC_count_TS559_Oct2020>=0))$mC_count_TS559_Oct2020, .99)[[1]]

    q_strain_dec <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r1)>=min_cov, !!as.name(m5C_count_r1)>=0) )[[m5C_count_r1]], .99)[[1]]
    
    q_strain_oct <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r2)>=min_cov, !!as.name(m5C_count_r2)>=0) )[[m5C_count_r2]], .99)[[1]]
    
    
    pooled_methylation_frequency_strain <- paste("pooled_methylation_frequency_", strain, sep='')
    
    data <- strain_cgmap %>%
        merge(y = TS559_cgmap, all = TRUE, 
             by= c("chromosome","position","reference_nucleotide","growth_phase","metabolic_condition")) %>%
        dplyr::select(-chromosome,-standard_devition) %>%
    
        # house keeping
        mutate(!!as.name(cov_r1) := replace_na(!!as.name(cov_r1),0)) %>%
        mutate(!!as.name(cov_r2) := replace_na(!!as.name(cov_r2),0)) %>%
        
        mutate(coverage_TS559_March2016 = replace_na(coverage_TS559_March2016,0)) %>%
        mutate(coverage_TS559_Dec2016 = replace_na(coverage_TS559_Dec2016,0)) %>%
        mutate(coverage_TS559_Oct2020 = replace_na(coverage_TS559_Oct2020,0)) %>%
        
        mutate(methylation_frequency_TS559_March2016 = na_if(methylation_frequency_TS559_March2016, "NA")) %>%    
        mutate(methylation_frequency_TS559_Dec2016 = na_if(methylation_frequency_TS559_Dec2016, "NA")) %>%    
        mutate(methylation_frequency_TS559_Oct2020 = na_if(methylation_frequency_TS559_Oct2020, "NA")) %>%    
        mutate(!!as.name(meth_freq_r1) := na_if(!!as.name(meth_freq_r1), "NA")) %>%  
        mutate(!!as.name(meth_freq_r2) := na_if(!!as.name(meth_freq_r2), "NA")) %>%    
        
        # identify high confidence and reproducible m5C sites
        rowwise() %>%
        mutate(hiconf_TS559 = detect_hiconf_3reps(
            f1=methylation_frequency_TS559_March2016,
            m1=mC_count_TS559_March2016,
            p1=q_TS559_march,
            c1=coverage_TS559_March2016,
            f2=methylation_frequency_TS559_Dec2016,
            m2=mC_count_TS559_Dec2016,
            p2=q_TS559_dec,
            c2=coverage_TS559_Dec2016,
            f3=methylation_frequency_TS559_Oct2020,
            m3=mC_count_TS559_Oct2020,
            p3=q_TS559_oct,
            c3=coverage_TS559_Oct2020,
            min_cov=min_cov,
            min_freq = minimum_freq) 
              ) %>%
        
        rowwise() %>%
        mutate(hiconf_strain = detect_hiconf_2reps(
                f1=!!as.name(meth_freq_r1),
                m1=!!as.name(m5C_count_r1),
                p1=q_strain_dec,
                c1=!!as.name(cov_r1),
                f2=!!as.name(meth_freq_r2),
                m2=!!as.name(m5C_count_r2),
                p2=q_strain_oct,
                c2=!!as.name(cov_r2), 
                min_cov= min_cov,
                min_freq = minimum_freq)
                 
                ) %>%
        # we are only interested in keeping sites were either the parent or deletion
        # strain have a m5C site
        filter(hiconf_TS559 == TRUE | hiconf_strain == TRUE) %>%
        filter(hiconf_TS559 != "NA") %>%
        filter(hiconf_strain != "NA") %>%
    
        # calculate a mean methylation frequency at each site by pooling replicates (round to 2 decimals)
        # calculate standard deviation of each site as well
        mutate(!!as.name(pooled_methylation_frequency_strain) := 
             ( !!as.name(m5C_count_r1) +  !!as.name(m5C_count_r2) ) / ( !!as.name(cov_r1) + !!as.name(cov_r2) ) ) %>%
        
        mutate(!!as.name(pooled_methylation_frequency_strain) := round(!!as.name(pooled_methylation_frequency_strain),2))%>%
        mutate(strain_sd = sd(c(!!as.name(meth_freq_r1),!!as.name(meth_freq_r2))))%>%
    
        mutate(pooled_methylation_frequency_TS559 =
               (mC_count_TS559_March2016 + mC_count_TS559_Dec2016 + mC_count_TS559_Oct2020) /
               (coverage_TS559_March2016+coverage_TS559_Dec2016+coverage_TS559_Oct2020)    ) %>%
        mutate(pooled_methylation_frequency_TS559=round(pooled_methylation_frequency_TS559,2)) %>%
        mutate(TS559_sd = sd(c(methylation_frequency_TS559_March2016,methylation_frequency_TS559_Dec2016,
                              methylation_frequency_TS559_Oct2020))) %>%
    

        # calculate fold change of methylaton frequency at each site between parent and deletion strain
        mutate(FoldChange = (!!as.name(pooled_methylation_frequency_strain)+.01) / 
                               (pooled_methylation_frequency_TS559+.01),
               log2FC = log2( (!!as.name(pooled_methylation_frequency_strain)+.01) / 
                               (pooled_methylation_frequency_TS559+.01) ) ) %>%
        
        filter(log2FC >= 1 | log2FC <= -1) %>%
        # add p vlaue
        rowwise() %>%
        mutate(pvalue = get_pvalue(
            s1_mC = !!as.name(m5C_count_r1),
            s1_cov = !!as.name(cov_r1), 
            s2_mC = !!as.name(m5C_count_r2),
            s2_cov = !!as.name(cov_r2),
            c1_mC = mC_count_TS559_March2016,
            c1_cov = coverage_TS559_March2016,
            c2_mC = mC_count_TS559_Dec2016,
            c2_cov = coverage_TS559_Dec2016,
            c3_mC = mC_count_TS559_Oct2020,
            c3_cov = coverage_TS559_Oct2020) ) %>%
        
        filter(pvalue <= 0.01)
        
        # adjust p values
        #adj_pvalue <- p.adjust(data$pvalue, method = 'fdr')
        #data <- cbind(data, adj_pvalue)
        data <- data %>% merge(y=annotation,by='position', all.x=TRUE) %>%
            dplyr::rename(TS559_position = position) %>%
            rowwise()%>%
            mutate(overlap = 
                   rm_overlap(!!as.name(pooled_methylation_frequency_strain), strain_sd,
                    pooled_methylation_frequency_TS559, TS559_sd )) %>%
           filter(overlap==FALSE) %>%
           dplyr::select(-overlap, -reproducible, -highly_reproducible) %>%
           arrange(pvalue)
    
    return(data)
}
defaultW <- getOption("warn")



enumerate_hiconf <- function(strain_cgmap, strain, min_cov=47){
    meth_freq_r1 <- paste("methylation_frequency_", strain, "_Dec2016", sep="")
    m5C_count_r1 <- paste("mC_count_", strain, "_Dec2016", sep="")
    cov_r1 <- paste("coverage_", strain, "_Dec2016", sep="")

    meth_freq_r2 <- paste("methylation_frequency_", strain, "_Oct2020", sep="")
    m5C_count_r2 <- paste("mC_count_", strain, "_Oct2020", sep="")
    cov_r2 <- paste("coverage_", strain, "_Oct2020", sep="")
    
    q_strain_dec <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r1)>=min_cov, !!as.name(m5C_count_r1)>=0) )[[m5C_count_r1]], .99)[[1]]
    
    q_strain_oct <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r2)>=min_cov, !!as.name(m5C_count_r2)>=0) )[[m5C_count_r2]], .99)[[1]]
    
    
    data <- strain_cgmap %>%
        dplyr::select(-chromosome, -reference_nucleotide,-growth_phase) %>%
    
        mutate(!!as.name(cov_r1) := replace_na(!!as.name(cov_r1),0)) %>%
        mutate(!!as.name(cov_r2) := replace_na(!!as.name(cov_r2),0)) %>%
        
        mutate(!!as.name(meth_freq_r1) := na_if(!!as.name(meth_freq_r1), "NA")) %>%  
        mutate(!!as.name(meth_freq_r2) := na_if(!!as.name(meth_freq_r2), "NA")) %>%    
        
        rowwise() %>%
        mutate(hiconf_strain = detect_hiconf_2reps(
                f1=!!as.name(meth_freq_r1),
                m1=!!as.name(m5C_count_r1),
                p1=q_strain_dec,
                c1=!!as.name(cov_r1),
                f2=!!as.name(meth_freq_r2),
                m2=!!as.name(m5C_count_r2),
                p2=q_strain_oct,
                c2=!!as.name(cov_r2), 
                min_cov=min_cov)
                 
                ) %>%
        filter(hiconf_strain == TRUE) %>%
        filter(!!as.name(cov_r1) >= min_cov, !!as.name(cov_r2) >= min_cov) %>%

        dplyr::rename(TS559_position = position) %>%
        
        mutate( 'm5C_cov/total_cov_Dec2016' := 
            paste(!!as.name(paste('mC_count_',strain,'_Dec2016',sep='')), 
                  '/', 
                  !!as.name(paste('coverage_',strain,'_Dec2016',sep='')) 
                 ) 
              ) %>%
        
        mutate( 'm5C_cov/total_cov_Oct2020' := 
            paste(!!as.name(paste('mC_count_',strain,'_Oct2020',sep='')), 
                  '/', 
                  !!as.name(paste('coverage_',strain,'_Oct2020',sep='')) 
                 ) 
              ) %>%
        dplyr::select(TS559_position, 
                  !!as.name(meth_freq_r1), 
                  'm5C_cov/total_cov_Dec2016',
                  !!as.name(meth_freq_r2), 
                  'm5C_cov/total_cov_Oct2020',
                     hiconf_strain) %>%

    dplyr::rename(!!paste(strain,'_hiconf',sep=''):= hiconf_strain )  %>%
    dplyr::rename(!!paste('m5C_cov/total_cov_',strain,'_rep1',sep=''):= 'm5C_cov/total_cov_Dec2016' )  %>%
    dplyr::rename(!!paste('m5C_cov/total_cov_',strain,'_rep2',sep=''):= 'm5C_cov/total_cov_Oct2020' )
    
    return(data)
}

get_quantiles <- function(strain_cgmap, strain, min_cov = 47){
    
    m5C_count_r1 <- paste("mC_count_", strain, "_Dec2016", sep="")
    cov_r1 <- paste("coverage_", strain, "_Dec2016", sep="")

    m5C_count_r2 <- paste("mC_count_", strain, "_Oct2020", sep="")
    cov_r2 <- paste("coverage_", strain, "_Oct2020", sep="")
    
    
    #get quantile values for m5C count
    q_strain_dec <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r1)>=min_cov, !!as.name(m5C_count_r1)>=0) )[[m5C_count_r1]], .99)[[1]]
    
    q_strain_oct <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r2)>=min_cov, !!as.name(m5C_count_r2)>=0) )[[m5C_count_r2]], .99)[[1]]

    return(c(q_strain_dec,q_strain_oct))
}

defaultW <- getOption("warn")
```

```r
###strain TK0008

CompareCGmap_1rep <- function(strain_cgmap, TS559_cgmap, strain, minimum_freq=.10, min_cov=47){
    
    meth_freq_r1 <- paste("methylation_frequency_", strain, "_Dec2016", sep="")
    m5C_count_r1 <- paste("mC_count_", strain, "_Dec2016", sep="")
    cov_r1 <- paste("coverage_", strain, "_Dec2016", sep="")

    meth_freq_r2 <- paste("methylation_frequency_", strain, "_Oct2020", sep="")
    m5C_count_r2 <- paste("mC_count_", strain, "_Oct2020", sep="")
    cov_r2 <- paste("coverage_", strain, "_Oct2020", sep="")
    
    
    #get quantile values for m5C count
    q_TS559_march <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_March2016>=min_cov, mC_count_TS559_March2016>=0))$mC_count_TS559_March2016, .99)[[1]]
    
    q_TS559_dec <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_Dec2016>=min_cov, mC_count_TS559_Dec2016>=0))$mC_count_TS559_Dec2016, .99)[[1]]
    
    q_TS559_oct <- quantile((TS559_cgmap%>%
        filter(coverage_TS559_Oct2020>=min_cov, mC_count_TS559_Oct2020>=0))$mC_count_TS559_Oct2020, .99)[[1]]

    q_strain_dec <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r1)>=min_cov, !!as.name(m5C_count_r1)>=0) )[[m5C_count_r1]], .99)[[1]]

    
    data <- strain_cgmap %>%
        merge(y = TS559_cgmap, all = TRUE, 
             by= c("chromosome","position","reference_nucleotide","growth_phase","metabolic_condition")) %>%
        dplyr::select(-chromosome,-standard_devition) %>%
    
        # house keeping
        mutate(!!as.name(cov_r1) := replace_na(!!as.name(cov_r1),0)) %>%
        mutate(!!as.name(cov_r2) := replace_na(!!as.name(cov_r2),0)) %>%
        
        mutate(coverage_TS559_March2016 = replace_na(coverage_TS559_March2016,0)) %>%
        mutate(coverage_TS559_Dec2016 = replace_na(coverage_TS559_Dec2016,0)) %>%
        mutate(coverage_TS559_Oct2020 = replace_na(coverage_TS559_Oct2020,0)) %>%
        
        mutate(methylation_frequency_TS559_March2016 = na_if(methylation_frequency_TS559_March2016, "NA")) %>%    
        mutate(methylation_frequency_TS559_Dec2016 = na_if(methylation_frequency_TS559_Dec2016, "NA")) %>%    
        mutate(methylation_frequency_TS559_Oct2020 = na_if(methylation_frequency_TS559_Oct2020, "NA")) %>%    
        
        mutate(!!as.name(meth_freq_r1) := na_if(!!as.name(meth_freq_r1), "NA")) %>%  
        
        dplyr::select(-!!as.name(meth_freq_r2),
                     -!!as.name(m5C_count_r2),
                     -!!as.name(cov_r2)) %>%
    
        # identify high confidence and reproducible m5C sites
        rowwise() %>%
        mutate(hiconf_TS559 = detect_hiconf_3reps(
            f1=methylation_frequency_TS559_March2016,
            m1=mC_count_TS559_March2016,
            p1=q_TS559_march,
            c1=coverage_TS559_March2016,
            f2=methylation_frequency_TS559_Dec2016,
            m2=mC_count_TS559_Dec2016,
            p2=q_TS559_dec,
            c2=coverage_TS559_Dec2016,
            f3=methylation_frequency_TS559_Oct2020,
            m3=mC_count_TS559_Oct2020,
            p3=q_TS559_oct,
            c3=coverage_TS559_Oct2020,
            min_cov=min_cov,
            min_freq = minimum_freq) ) %>%
        
        rowwise() %>%
    
    mutate(hiconf_strain = detect_hiconf_1rep(
                f1=!!as.name(meth_freq_r1),
                m1=!!as.name(m5C_count_r1),
                p1=q_strain_dec,
                c1=!!as.name(cov_r1), 
                min_cov=min_cov) ) %>%
           
           
        # we are only interested in keeping sites were either the parent or deletion
        # strain have a m5C site
        filter(hiconf_TS559 == TRUE | hiconf_strain == TRUE) %>%
        mutate(hiconf_TS559 = na_if(hiconf_TS559, "NA")) %>%    
        mutate(hiconf_TS559 = na_if(hiconf_TS559, "NA")) %>%    
        filter(hiconf_TS559 != "NA" & hiconf_strain != "NA") %>%
    
        # calculate a mean methylation frequency at each site by pooling replicates (round to 2 decimals)
    
        mutate(pooled_methylation_frequency_TS559 =
               (mC_count_TS559_March2016 + mC_count_TS559_Dec2016 + mC_count_TS559_Oct2020) /
               (coverage_TS559_March2016+coverage_TS559_Dec2016+coverage_TS559_Oct2020)    ) %>%
        mutate(pooled_methylation_frequency_TS559=round(pooled_methylation_frequency_TS559,2)) %>%
    
        # caclutate standard deviation of each site as well
        mutate(strain_sd = 0)%>%

        mutate(TS559_sd = sd(c(methylation_frequency_TS559_March2016,methylation_frequency_TS559_Dec2016,
                              methylation_frequency_TS559_Oct2020))) %>%
        
        
        # calcualte fold change of methylaton frequency at each site between parent and deletion strain
        mutate(FoldChange = (!!as.name(meth_freq_r1) / pooled_methylation_frequency_TS559 ),
            log2FC = log2(!!as.name(meth_freq_r1) / pooled_methylation_frequency_TS559 ) ) %>%
        
        filter(log2FC >= 1 | log2FC <= -1) %>%

        # add p vlaue
        rowwise() %>%
        mutate(pvalue = get_pvalue_1rep(
            s1_mC = !!as.name(m5C_count_r1),
            s1_cov = !!as.name(cov_r1), 
            c1_mC = mC_count_TS559_March2016,
            c1_cov = coverage_TS559_March2016,
            c2_mC = mC_count_TS559_Dec2016,
            c2_cov = coverage_TS559_Dec2016,
            c3_mC = mC_count_TS559_Oct2020,
            c3_cov = coverage_TS559_Oct2020) ) %>%
        
        filter(pvalue <= 0.05)
        
        data <- data %>% merge(y=annotation,by='position', all.x=TRUE) %>%
            dplyr::rename(TS559_position = position) %>%
            rowwise()%>%
            mutate(overlap = 
                   rm_overlap(!!as.name(meth_freq_r1), strain_sd,
                    pooled_methylation_frequency_TS559, TS559_sd )) %>%
           filter(overlap==FALSE) %>%
           dplyr::select(-reproducible, -highly_reproducible) %>%
           arrange(log2FC)
    return(data)
}

get_quantiles_1rep <- function(strain_cgmap, strain, min_cov = 47){
    
    m5C_count_r1 <- paste("mC_count_", strain, "_Dec2016", sep="")
    cov_r1 <- paste("coverage_", strain, "_Dec2016", sep="")

    
    #get quantile values for m5C count
    q_strain_dec <- quantile(
        (strain_cgmap %>%
        filter(!!as.name(cov_r1)>=min_cov, !!as.name(m5C_count_r1)>=0) )[[m5C_count_r1]], .99)[[1]]

    return(c(q_strain_dec))
}

defaultW <- getOption("warn")
options(warn = -1)


#load CGmaps into session
TK0008exoS_CGmap_Dec2016 <-read.delim("../cgmaps/TK0008exoS_totalRNA_Dec2016.CGmap", sep='\t', header=FALSE,col.names=cols) %>%
    mutate(strain="TK0008", growth_phase='exoponential', metabolic_condition="S", replicate="Dec2016")

TK0008statS_CGmap_Dec2016 <-read.delim("../cgmaps/TK0008statS_totalRNA_Dec2016.CGmap", sep='\t', header=FALSE,col.names=cols) %>%
    mutate(strain="TK0008", growth_phase='stationary', metabolic_condition="S", replicate="Dec2016")

TK0008exoS_CGmap_Oct2020 <-read.delim("../cgmaps/TK0008exoS_totalRNA_Oct2020.CGmap", sep='\t', header=FALSE,col.names=cols) %>%
    mutate(strain="TK0008", growth_phase='exoponential', metabolic_condition="S", replicate="Oct2020")

TK0008statS_CGmap_Oct2020 <-read.delim("../cgmaps/TK0008statS_totalRNA_Oct2020.CGmap", sep='\t', header=FALSE,col.names=cols) %>%
    mutate(strain="TK0008", growth_phase='stationary', metabolic_condition="S", replicate="Oct2020")

# process raw CGmaps
TK0008exoS_Dec2016_df <- processCGmap(TK0008exoS_CGmap_Dec2016)
TK0008exoS_Oct2020_df <- processCGmap(TK0008exoS_CGmap_Oct2020)
TK0008statS_Dec2016_df <- processCGmap(TK0008statS_CGmap_Dec2016)
TK0008statS_Oct2020_df <- processCGmap(TK0008statS_CGmap_Oct2020)

#merge reps
TK0008exoS <- merge_del_reps(x=TK0008exoS_Dec2016_df, y=TK0008exoS_Oct2020_df, strain='TK0008')
TK0008statS <- merge_del_reps(x=TK0008statS_Dec2016_df, y=TK0008statS_Oct2020_df,  strain='TK0008')



eTK0008exoS <- enumerate_hiconf(TK0008exoS, "TK0008")
eTK0008statS <- enumerate_hiconf(TK0008statS, "TK0008")


options(warn = defaultW)
```

```r
options(repr.plot.width = 2, repr.plot.height = 2)

lin_reg <- function(df, strain, phase) {
    x = paste("methylation_frequency_", strain, "_Dec2016", sep='')
    y = paste("methylation_frequency_", strain, "_Oct2020", sep='')
    title = paste("Δ", strain, sep='')
    if(phase == 'exoS'){pc = 'grey90'} else if(phase == 'statS'){pc='red'}
    
ggplot(df, aes(x=!!as.name(x), y=!!as.name(y)))+
    geom_point() +
    geom_smooth(method = "glm", color='blue3') +
    theme_bw() +
    scale_x_continuous(limits=c(0,1.00)) +
    scale_y_continuous(limits=c(0,1.00)) +
    labs(x="replicate 1", y="replicate 2", title = title)+
    theme(axis.text = element_text(color='black'),
          plot.title = element_text(hjust = 0.5),
          panel.background = element_rect(fill = pc)
    )
  
}

lin_reg(eTK2304exoS, "TK2304", "exoS")
```
